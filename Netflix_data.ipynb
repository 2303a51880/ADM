```python
import pandas as pd
url='/content/netflix_titles.csv'
data = pd.read_csv(url)
print(data.describe())
print(data)
```

```python
import matplotlib.pyplot as plt
plt.hist(data['release_year'].dropna(),bins=20)
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.title('Frequency of Released Year')
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(data['release_year'].dropna())
plt.ylabel('Released Year')
plt.title('Boxplot of Released Year')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(data[['release_year', 'country', 'date_added', 'type']].dropna())
plt.show()
```

```python
sns.countplot(x='release_year',data=data,palette='Set2')
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.title('Frequency of Released Year')
plt.show()
```

```python
sns.violinplot(x='release_year', data=data, palette='Set2')
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.title('Frequency of Released Year')
plt.show()
```
