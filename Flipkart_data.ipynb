```python
import pandas as pd
data=pd.read_csv('/content/flipkart_com-ecommerce_sample.csv')
print(data.describe())
```

```python
import matplotlib.pyplot as plt
plt.hist(data['retail_price'].dropna(),bins=20)
plt.xlabel('Retail Price')
plt.ylabel('Frequency')
plt.title('Frequency of Released Year')
plt.show()
```

```python
import seaborn as sns

sns.pairplot(data[['retail_price','discounted_price']].dropna())
plt.show()
```

```python
# Correlation heatmap
correlation_matrix = data[['retail_price','discounted_price']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='retail_price',data=data,palette='Set2')
plt.xlabel('Released Year')
plt.ylabel('Frequency')
plt.title('Frequency of Released Year')
plt.show()
```

```python
sns.violinplot(x='retail_price',y='discounted_price', data=data, palette='Set2')
plt.xlabel('retail price')
plt.ylabel('discounted price')
plt.title('retail to discounted price ')
plt.show()
```
